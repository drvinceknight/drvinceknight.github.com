```python
import pathlib
import yaml

students_year = {}
students_tags = {}
students = set()
there = pathlib.Path("../src/data/StudentProject/")
for file in there.glob("*.yml"):
    data = yaml.safe_load(file.read_text())
    student = data["student"]
    students_year[student] = data["end_date"]
    if student in students_tags:
        students_tags[student].append(data["type"])
    else:
        students_tags[student] = [data["type"]]

    students = students | {student}
```

```python
here = pathlib.Path("./")


for file in here.glob("*.md"):
    contents = file.read_text()
    if (student:=file.stem) in students:
        contents = file.read_text()

        contents = contents.replace("tag", "category")
        contents = contents.replace("name", f"year: {students_year[student]}\nname")
        tags_string = "tags:\n"
        for tag in set(students_tags[student]):
            tags_string += f"  - {tag}\n"
        contents = contents.replace("name", f"{tags_string}name")

        print(contents)
        with open(f"{file.stem}.md", "w") as f:
            f.write(contents)
```
